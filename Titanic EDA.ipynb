```python
# Creating new column family size 
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
sns.barplot(df['FamilySize'].value_counts())
```

```python
# Creating family type column - alone, small, big
def transform_family_size(size):
    if size == 1:
        return 'alone'
    elif size > 1 and size <5:
        return 'small'
    else:
        return 'big'


df['FamilyType'] = df['FamilySize'].apply(transform_family_size)
```

```python
df['FamilyType'].value_counts().plot(kind = 'pie', autopct = '%1.1f%%')
```

```python
df['FamilyType'].isna().sum()
```

```python
# creating new columns age group
def find_age_group(age):
    if age <= 12:
        return 'child'
    elif age >13 and age <= 20:
        return 'teen'
    elif age > 21 and age < 64:
        return 'adults'
    elif age >= 64:
        return 'old'

```

```python
df['AgeGroup'] = df['Age'].apply(find_age_group)
```
